```python
import cbpro
public_client = cbpro.PublicClient()
```

```python
# get products
products = pd.DataFrame(public_client.get_products())
products
```

```python
products.columns
```

```python
products.id.to_list()
```

```python
public_client.get_product_order_book('BTC-USD', level=1)
```

```python
public_client.get_product_ticker('BTC-USD')
```

```python
price = pd.DataFrame(public_client.get_product_historic_rates('BTC-USD', start='2022-09-18 00:00:00',end='2022-09-18 01:00:00',  granularity=60),  #
                     columns=['time', 'low', 'high', 'open', 'close', 'volume'])
price
```

```python
price.time = pd.to_datetime(price.time, unit='s')
price
```

```python
price.time.apply(pd.Timestamp, unit='s', tz='US/EASTERN')
```
